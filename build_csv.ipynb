```python
import pandas as pd
import os
import csv

# Path to the directory containing WAV files
wav_path = '/content/drive/My Drive/ML-PROJECT/sounds/full4' # MODIFY HERE

# List of files in the specified directory
wav_files = [f for f in os.listdir(wav_path) if f.endswith('.wav')]

# Create a temporary DataFrame with new data
new_data = pd.DataFrame(wav_files, columns=['file_name'])

# Add additional columns to the DataFrame
new_data['target'] = [characters[i*4:(i+1)*4] for i in range(len(new_data))]  # Split the string every 4 characters
new_data['target'] = new_data['target'].astype(str)  # Ensure the data in the 'target' column is treated as strings
new_data['dataset_type'] = 'test'  # MODIFY HERE

# Path to the existing CSV file
csv_path = '/content/drive/My Drive/ML-PROJECT/information.csv'

# Check if the file exists. If not, create a file with headers; otherwise, append the data.
if not os.path.isfile(csv_path):
    new_data.to_csv(csv_path, index=False, quoting=csv.QUOTE_NONNUMERIC)
else:
    new_data.to_csv(csv_path, mode='a', header=False, index=False, quoting=csv.QUOTE_NONNUMERIC)
```
